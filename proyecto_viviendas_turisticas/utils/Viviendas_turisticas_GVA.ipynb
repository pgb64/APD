```python
# Unimos todos los csv
archivos = glob.glob("csv_municipios/*.csv")

lista_df = []

for f in archivos:
    municipio = os.path.basename(f).replace(".csv", "")

    try:
        df = pd.read_csv(f, sep=None, engine="python")  # AUTODETECCIÓN DEL SEPARADOR
        df["municipio"] = municipio  # para saber de dónde viene cada fila
        lista_df.append(df)

    except Exception as e:
        print(f"Error leyendo {f}: {e}")

df_final = pd.concat(lista_df, ignore_index=True)

df_final.to_csv("csv_municipios/viviendas_sucias.csv", index=False, encoding="utf-8-sig")

print("✔ Archivo final creado correctamente: viviendas_sucias.csv")
```
